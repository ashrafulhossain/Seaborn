```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
var1 = sns.load_dataset("tips")
var1
```

# Count Plot

```python
sns.countplot(x='sex', data=var1, hue='smoker')
plt.show()
```

```python
var1
```

# Violine Plot

```python
sns.violinplot(x='day', y='total_bill', data=var1, palette='Dark2')
plt.show()
```

# Violine Plot

```python
sns.violinplot(x='day', y='total_bill', data=var1, hue= 'time')
plt.show()
```

```python
sns.violinplot(x='time', y='total_bill', data=var1, order=['Dinner', 'Lunch'], palette='Dark2')
plt.show()
```

# Pair Plot

```python
var1
```

```python
sns.pairplot(var1, hue='sex')
plt.show()
```

```python

sns.pairplot(var1, vars=['total_bill', 'tip'], hue='sex', markers=['o', '>'])

# Show the plot
plt.show()
```

```python
var1
```

# Strip Plot


```python
sns.stripplot(x='day', y='total_bill', data= var1, hue = 'sex', palette='bright')
plt.show()
```

```python
var1
```

# Box Plot

```python
sns.set(style='whitegrid')
sns.boxplot(x='day', y='total_bill', data=var1, hue='sex', order=['Fri', 'Sun', 'Thur', 'Sat'], showmeans=True, linewidth=1.5)
plt.show()

```

```python
var1
```

# Cat Plot


```python
sns.catplot(x='size', y='tip', data=var1, hue='sex', kind='point') ## here use can "kind" parametar then we are create to different types of plot look like boxplot,barplot etc,
plt.show()
```

```python
var1
```

# Facet Grid


```python
fg = sns.FacetGrid(var1, col="day", hue='sex')
fg.map(plt.bar, "total_bill", "tip").add_legend()
plt.show()
```
